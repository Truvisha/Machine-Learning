```python
#Edge detection
import cv2
import matplotlib.pyplot as plt

#Loading an image
img=cv2.imread('au.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)#Edge detection works on intensity value,not on color.So BGR2GRAY is used

#sobel(src,ddepth,x,y,kernel_size)
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,10)
sobely=cv2.Sobel(img,cv2.CV_64F,1,0,10)
sobel_combined=cv2.magnitude(sobelx,sobely)#Magnitude combines storng edges from sobel-x and sobel-y

#ploting the image
plt.figure(figsize=(10,6))
plt.subplot(1,4,1),plt.imshow(img),plt.title('Source image'),plt.axis('off')
plt.subplot(1,4,2),plt.imshow(sobelx,cmap='gray'),plt.title('Sobel-X'),plt.axis('off')
plt.subplot(1,4,3),plt.imshow(sobely,cmap='gray'),plt.title('Sobel-Y'),plt.axis('off')
plt.subplot(1,4,4),plt.imshow(sobel_combined,cmap='gray'),plt.title('Sobel-Combined'),plt.axis('off')
           
```

```python
#Edge detection
import cv2
import matplotlib.pyplot as plt

#Loading an image
img=cv2.imread('au.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)#Edge detection works on intensity value,not on color.So BGR2GRAY is used

#Blurring an image using Gaussian blur
gauss=cv2.GaussianBlur(img,(3,3),0)

#Laplacian(src,ddepth,kernel_size)
laplacian=cv2.Laplacian(img,cv2.CV_64F,ksize=3)
laplacian_after_blur=cv2.Laplacian(gauss,cv2.CV_64F,ksize=3)


#ploting the image
plt.figure(figsize=(10,6))
plt.subplot(1,3,1),plt.imshow(img),plt.title('Source image'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(laplacian,cmap='gray'),plt.title('Lap'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(laplacian_after_blur,cmap='gray'),plt.title('Lap_Gray'),plt.axis('off')
```

```python
#Edge detection
import cv2
import matplotlib.pyplot as plt

#Loading an image
img=cv2.imread('au.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)#Edge detection works on intensity value,not on color.So BGR2GRAY is used

#Blurring an image using Gaussian blur
gauss=cv2.GaussianBlur(img,(3,3),0)

#canny(src,threshold1,threshold2)
canny=cv2.Canny(img,threshold1=20,threshold2=500)
canny_after_blur=cv2.Canny(gauss,threshold1=20,threshold2=150)

#ploting the image
plt.figure(figsize=(10,6))
plt.subplot(1,3,1),plt.imshow(img),plt.title('Source image'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(canny,cmap='gray'),plt.title('canny'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(canny_after_blur,cmap='gray'),plt.title('canny_after_blur'),plt.axis('off')
```
