```python
#Loading the image
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('neu.jpg')
plt.imshow(img)
```

```python
#convert image from BGR to RGB
img=cv2.imread('neu.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
